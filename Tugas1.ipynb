```python
import pandas as pd
heart = pd.read_csv('/content/heart.csv')
```

```python
heart.info()
```

```python
heart.head(10)
```

```python
heart.drop('age',axis=1, inplace=True)
```

```python
heart.head()
```

```python
#memisakhan atribut dengan label dimana x itu atribut dan y itu label
X = heart[['anaemia','creatinine_phosphokinase','diabetes','ejection_fraction','high_blood_pressure','platelets','serum_creatinine','serum_sodium','sex','smoking','time']]
y = heart['DEATH_EVENT']


#membagi dataset menjadi data latih dan data uji
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.1, random_state=123)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Membuat model Random Forest
forest_model = RandomForestClassifier()

# Melatih model menggunakan data latih dengan fungsi fit
forest_model = forest_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

# Menggunakan model Random Forest untuk membuat prediksi
y_pred = forest_model.predict(X_test)

# Menghitung akurasi
accuracy = round(accuracy_score(y_test, y_pred), 3)

print('Accuracy:', accuracy)
```

```python
# Prediksi menggunakan model Random Forest
prediction = forest_model.predict([[0,582,0,2,1,265,1.9,130,1,0,4]])

print(prediction[0])
```

```python
prediction = forest_model.predict([[1,160,1,20,0,327,2.7,116,0,0,8]])

print(prediction[0])
```

```python
# Melihat class names dari model Random Forest
print(forest_model.classes_)
```

```python
from sklearn.tree import export_graphviz

# Pilih salah satu pohon dari model Random Forest
tree = forest_model.estimators_[0]

# Menentukan nama kelas sesuai dengan yang digunakan dalam model
class_names = [str(cls) for cls in forest_model.classes_]

export_graphviz(
    tree,
    out_file="heart_random_forest.dot",
    feature_names=['anaemia','reatinine_phosphokinase','diabetes','ejection_fraction','high_blood_pressure','platelets','serum_creatinine','serum_sodium','sex','smoking','time'],
    class_names=class_names,
    rounded=True,
    filled=True
)
```
